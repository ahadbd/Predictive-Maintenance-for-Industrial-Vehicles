```python
import pandas as pd
import matplotlib.pyplot as plt

# Load results
df = pd.read_csv("../data/predictive_maintenance_results.csv")

# Separate normal vs maintenance required
normal = df[df['maintenance_required'] == 0]
maintenance = df[df['maintenance_required'] == 1]

# ------------------------
# Scatter 1: Hours since last maintenance vs Engine temp
# ------------------------
plt.figure(figsize=(10,6))
plt.scatter(normal['hours_since_last_maintenance'], normal['temperature'], c='blue', label='No Maintenance', alpha=0.6)
plt.scatter(maintenance['hours_since_last_maintenance'], maintenance['temperature'], c='red', label='Maintenance Required', alpha=0.8)
plt.xlabel('Hours since Last Maintenance')
plt.ylabel('Engine Temperature (°C)')
plt.title('Predictive Maintenance: Engine Temperature vs Hours')
plt.legend()
plt.grid(True)
plt.show()

# ------------------------
# Scatter 2 (Bonus): Fuel Rate vs Idle Time
# ------------------------
plt.figure(figsize=(10,6))
plt.scatter(normal['fuel_rate'], normal['idle_time'], c='blue', label='No Maintenance', alpha=0.6)
plt.scatter(maintenance['fuel_rate'], maintenance['idle_time'], c='red', label='Maintenance Required', alpha=0.8)
plt.xlabel('Fuel Rate (L/h)')
plt.ylabel('Idle Time (s)')
plt.title('Predictive Maintenance: Fuel Rate vs Idle Time')
plt.legend()
plt.grid(True)
plt.show()

# ------------------------
# Summary Metrics
# ------------------------
total_points = len(df)
maintenance_count = len(maintenance)
normal_count = len(normal)
maintenance_percentage = maintenance_count / total_points * 100

print(f"Total data points: {total_points}")
print(f"No maintenance required: {normal_count}")
print(f"Maintenance required: {maintenance_count} ({maintenance_percentage:.2f}%)")
```
